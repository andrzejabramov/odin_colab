```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/drive/MyDrive/data/winequality-red.csv", sep=";")
```

```python
df
```

```python
df.columns
```

```python
sns.barplot(df['quality'].value_counts())
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(df['quality'].value_counts(normalize=True))
plt.title("Распределение качества")
plt.xlabel("Качество")
plt.ylabel("Количество")
plt.show()
```

```python
sns.kdeplot(df, x="quality")
```

```python
plt.figure(figsize=(12, 6))
#настройка цвета гистограммы color='skyblue' настройка цвета границ edgecolor='black'
#kde=True отображает график плотности распределения на гистограмме, самый простой способ их объединить
sns.histplot(df['alcohol'], kde=True, bins=30,
             alpha=0.5, color='skyblue', edgecolor='black')
#линия показывающая среднее значение
plt.axvline(df['alcohol'].mean(), color='green', linestyle='--',
            label=f'Среднее: {df['alcohol'].mean():.2f}')

plt.title(f'Распределение признака {'alcohol'}')
plt.xlabel('alcohol')
plt.ylabel('Плотность')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
