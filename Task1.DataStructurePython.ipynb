Задание 1: используем функции списков min и max. Обрабатываем исключение, если список пуст.

```python
def min_max(numbers):
    if not numbers:
        raise ValueError("There are no elements")
    return [min(numbers), max(numbers)]

print(min_max([3, 1, 4, 1, 5, 9, 2, 6]))
```

Задание 2. Возвращаем список чисел из `numbers`, которые больше `threshold`. Проверку на наличие элементов в списке делать не стал.

```python
def filter_greater(numbers, threshold):
    return [num for num in numbers if num > threshold]

print(filter_greater([1, 5, 3, 8, 2, 9, 0], 4))
```

Второй вариант с лямбда:

```python
def filter_greater(numbers, threshold):
    return list(filter(lambda x: x > threshold, numbers))

print(filter_greater([1, 5, 3, 8, 2, 9, 0], 4))
```

1)конфеты:

```python
def calculate_candy_price():
    # Запрашиваем вес у пользователя
    weight_input = input("Введите вес конфет (в килограммах): ")

    # Проверяем, можно ли преобразовать в число
    try:
        weight = float(weight_input)
    except ValueError:
        print("Ошибка: введите корректное число.")
        return

    # Проверяем минимальный вес
    if weight < 0.5:
        print("Минимальный вес покупки — 500 грамм (0.5 кг).")
        return

    # Определяем цену за килограмм в зависимости от веса
    if weight > 1.5:
        price_per_kg = 200
    else:
        price_per_kg = 250

    # Считаем общую стоимость
    total_cost = weight * price_per_kg

    # Выводим результат
    print(f"Цена за килограмм: {price_per_kg} руб.")
    print(f"Вес: {weight} кг")
    print(f"Итоговая стоимость: {total_cost:.2f} руб.")


calculate_candy_price()
```

Задание про кредит. В задании указано: Пользователь должен ввести __ТРИ__ значения: срок кредита в месяцах и сумму (получается два - ошибка?), а вывести ежемесячный платеж

```python
def calculate_loan():
    # Ввод данных
    amount_input = input("Введите сумму кредита (в рублях): ")
    term_input = input("Введите срок кредита (в месяцах): ")

    # Проверка на корректность ввода
    try:
        amount = float(amount_input)
        term = int(term_input)  # Срок обычно целое число
    except ValueError:
        print("Ошибка: введите корректные числовые значения.")
        return

    # Проверка на положительные значения
    if amount <= 0:
        print("Сумма кредита должна быть больше нуля.")
        return
    if term <= 0:
        print("Срок кредита должен быть больше нуля.")
        return

    # Логика по сроку
    if term > 12:
        print("Кредит не выдаётся: максимальный срок — 12 месяцев.")
        return
    elif term <= 6:
        interest_rate = 0.05  # 5%
    else:  # от 7 до 12
        interest_rate = 0.10  # 10%

    # Общая сумма к возврату (с переплатой)
    total_amount = amount * (1 + interest_rate)

    # Ежемесячный платёж
    monthly_payment = total_amount / term

    # Вывод результата
    print(f"Сумма кредита: {amount:.2f} руб.")
    print(f"Срок: {term} месяцев")
    print(f"Процентная ставка: {interest_rate * 100}%")
    print(f"Общая сумма к возврату: {total_amount:.2f} руб.")
    print(f"Ежемесячный платёж: {monthly_payment:.2f} руб.")


calculate_loan()
```

1) Циклы - деление без остатка:

```python
def print_multiples(N, K):

    # Выводит числа от 0 до N, кратные K.

    if math.floor(K) == 0:
        print("Ошибка: K не может быть нулём.")
        return

    if not isinstance(N, (int, float)) or not isinstance(K, (int, float)):
        print("Ошибка: N и K должны быть числами.")
        return

    # Приводим к целым, если нужно
    N = int(N)
    K = int(K)

    step = abs(K)
    print(f"Числа от 0 до {N}, кратные {K}:")

    for i in range(0, N + 1, step):
        print(i)


print_multiples(20, 3)
```

2) Все делители числа Х:

```python
import math

def print_divisors_fast(X):
    # Быстрый способ найти делители (кроме 1 и X).
    if not isinstance(X, int):
        print("Ошибка: X должно быть целым числом.")
        return

    num = abs(X)
    if num < 2:
        print(f"У числа {X} нет делителей.")
        return

    divisors = set()

    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            divisors.add(i)
            if i != num // i:  # избегаем дублирования (например, 5 и 5 для 25)
                divisors.add(num // i)

    divisors = sorted(divisors)

    if divisors:
        print(f"Делители числа {X} (кроме 1 и {X}): {divisors}")
    else:
        print(f"Число {X} — простое.")


print_divisors_fast(30)
```

3) Возведение в степень циклом:

```python
def power_by_loop(base, exponent):
    if not isinstance(exponent, int):
        print("Ошибка: степень должна быть целым числом.")
        return None

    if base == 0 and exponent < 0:
        print("Ошибка: ноль в отрицательной степени не определён.")
        return None

    if exponent == 0:
        result = 1
    elif exponent > 0:
        result = 1
        for _ in range(exponent):
            result *= base
    else:  # exponent < 0
        result = 1
        for _ in range(-exponent):
            result *= base
        result = 1 / result

    return f"{base} в степени {exponent} = {result}"


power_by_loop(2, 5)
```

```python
power_by_loop(3, 0)
```

```python
power_by_loop(2, -3)
```

```python
power_by_loop(5.5, 2)
```

4) простое число

```python
def is_prime(N):
    if not isinstance(N, int):
        return False
    if N <= 1:
        return False
    if N == 2:
        return True
    if N % 2 == 0:
        return False

    i = 3
    while i * i <= N:  # вместо sqrt(N)
        if N % i == 0:
            return False
        i += 2
    return True


print(is_prime(2))
print(is_prime(3))
print(is_prime(4))
print(is_prime(17))
print(is_prime(25))
print(is_prime(1))
print(is_prime(-5))
```
