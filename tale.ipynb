```python
skazki_link = rq.get("https://nukadeti.ru/skazki")
print(skazki_link.status_code)
```

```python
soup_skazki = bs4.BeautifulSoup(skazki_link.text, "html.parser")
soup_skazki
```

```python
addresslist = []

for tag in soup_skazki.find_all('a', class_='title'):
  addresslist.append('https://nukadeti.ru' + tag.get('href'))
addresslist
```

```python
counter = 0

while(counter < 5):
  try:
    skaska_link = rq.get(random.choice(addresslist))
    newSoup = bs4.BeautifulSoup(skaska_link.text, 'html.parser')
    newSoup.find("div", class_="tale-text").text
    break
  except Exception as e:
    counter += 1
```
